```python
from IPython.display import display
import numpy as np
import pandas as pd
import pyvinecopulib as pv
```

```python
n = 1000
d = 3
structure = np.matrix([[2,2,2],[3,3,np.nan],[1,np.nan,np.nan]])
pair_copulas = [[
    pv.Bicop(family=pv.BicopFamily.gaussian, parameters=[[.75]]),
    pv.Bicop(family=pv.BicopFamily.gaussian, parameters=[[.25]])
],
[
    pv.Bicop(family=pv.BicopFamily.gaussian, parameters=[[.25]])
]]
cop = pv.Vinecop(structure, pair_copulas)
U = cop.simulate(n=n)
```
