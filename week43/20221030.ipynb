```python
import numpy as np
import pyvinecopulib as pv
from matplotlib import pyplot as plt

kt = 0.8
n = 1000

family_lst = [pv.BicopFamily.clayton, pv.BicopFamily.frank, pv.BicopFamily.gaussian]
family_lab = ['Clayton', 'Frank', 'Gaussian']
fig, axes = plt.subplots(1,3, sharex=True, sharey=True,figsize=(9,3))
axes = axes.flatten()

for i, family in enumerate(family_lst):
    parameters = pv.Bicop(family=family).tau_to_parameters(kt)
    bcop = pv.Bicop(family=family, parameters=parameters)
    X = bcop.simulate(n)
    axes[i].scatter(X[:,0],X[:,1],marker='.', color='k', alpha=0.8)
    axes[i].set_xlim([0,1])
    axes[i].set_ylim([0,1])
    axes[i].set_title(family_lab[i])
plt.tight_layout()
plt.savefig('bivarcopula.png', dpi=300)
```
